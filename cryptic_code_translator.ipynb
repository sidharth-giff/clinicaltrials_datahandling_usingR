```python
print("cryptic translator")

translator = {
    " ": "//",
    "a": "?",  "b": ">",  "c": "<",  "d": "!",  "e": "@",
    "f": "}",  "g": ":",  "h": "+",  "i": "#",  "j": "$",
    "k": "%",  "l": "^",  "m": "&",  "n": "*",  "o": "()",
    "p": "__", "q": "==", "r": "}}", "s": "##", "t": "@@",
    "u": "!!", "v": "^^", "w": "**", "x": "??", "y": "$$", "z": "%%"
}


user_input = input("enter a text to translate: ").lower()

translated_message = ""

for char in user_input:
    if char in translator:
        translated_message += translator[char]
    else:
        translated_message += char


print("Translated text:", translated_message)
```
